```python
shipping = data.drop(['SeniorCitizen', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod'],axis=1)
```

```python
shipping.head()
```

```python
correlation = shipping.corr(numeric_only=True)
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(shipping)
```

```python
sns.relplot(x='tenure', y='TotalCharges', hue='gender', data=shipping)
```
